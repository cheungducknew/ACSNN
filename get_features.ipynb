# see kpgt features

## trans 2 key-value

```python
import numpy as np
import pandas as pd

dataset = "thrombin"
base_dir = "~/autodl-fs/KPGT/datasets"
data_dir = f"{base_dir}/{dataset}_mmps/{dataset}_mmps.csv"
feature_dir = f"../KPGT/datasets/{dataset}_mmps/kpgt_base.npz"
save_dir = f"data/{dataset}_kpgt.npz"

df = pd.read_csv(data_dir)
feature = np.load(feature_dir)
df.head()
```

```python
fps = feature['fps']
len(fps),len(df),len(fps[0]),fps
```

```python
smiles_features_dict = {}
smiles_list = df['smiles'].tolist()
for i, smiles in enumerate(smiles_list):
    smiles_features_dict[smiles] = fps[i]

np.savez_compressed(save_dir, **smiles_features_dict)
```
